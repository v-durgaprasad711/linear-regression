```python
import numpy as np

class LinearRegression():
  def __init__(self,Learning_rate,no_of_iterators):
    self.learning_rate=Learning_rate
    self.no_of_iterators=no_of_iterators
  def fit(self,x,y):
    self.m,self.n=x.shape
    self.w=np.zeros(self.n)
    self.b=0
    self.x=x
    self.y=y

    for i in range(self.no_of_iterators):
      self.upgrade_weights()

  def upgrade_weights(self):
    y_prediction=self.predict(self.x)
    dw=-2*(self.x.T).dot(self.y-y_prediction)/self.m
    db=(-2*np.sum(self.y-y_prediction))/self.m

    self.w=self.w-self.learning_rate*dw
    self.b=self.b-self.learning_rate*db


  def predict(self,x):
    return x.dot(self.w)+self.b
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

salary_data=pd.read_csv("/content/salary_data.csv")
x=salary_data.iloc[:,:-1].values
y=salary_data.iloc[:,1].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=2)
```

```python
model=LinearRegression(Learning_rate=0.02,no_of_iterators=1000)
```

```python
model.fit(x_train,y_train)
```

```python
print("weights",model.w)
```

```python
print("bias",model.b)
```

```python
test_data_prediction=model.predict(x_test)
print(x_test,test_data_prediction)
```

```python
plt.scatter(x_test,y_test,color="red")
plt.plot(x_test,test_data_prediction,color="blue")
plt.show()
```
